```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
from skimage import io, color
from skimage import exposure
```

```python
from skimage import data
```

```python
img = data.astronaut()
```

```python
plt.imshow(img)
```

```python
img[:,:,1]
```

```python
img = color.rgb2gray(img)
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
kernelv = np.array([[-1/2,0,1/2.0]])
img.ndim
```

```python
vedges = scipy.signal.convolve2d (img, kernelv, 'valid')
```

```python
plt.imshow(vedges, cmap='gray')
```

```python
kernelh = np.array([[-1/2,0,1/2.0]]).T
```

```python
hedges = scipy.signal.convolve2d (img, kernelh, 'valid')
```

```python
plt.imshow(hedges, cmap='gray')
```

```python
laplacian = np.array([[0,-1,0],
                      [-1,4,-1],
                      [0,-1,0]])
```

```python
edges = scipy.signal.convolve2d (img, laplacian, 'valid')
```

```python
plt.imshow(edges, cmap='gray')
```

```python
bkernel = np.ones((7,7))/9
bkernel
```

```python
blurr_img = scipy.signal.convolve2d (img, bkernel, 'valid')
```

```python
plt.imshow(blurr_img, cmap='gray')
```
